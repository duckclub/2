1. Determine the value k, the number of clusters
2. We have to assign each of the n examples to one of the k
clusters
3. We have to minimise the differences within each cluster
4. We have to maximise the differences between the clusters
5. Identify the centre elements in each cluster
6. Initially, they are chosen randomly
7. Using the Euclidean distance function from the centre, identify
and assign nearest examples to each cluster
8. Update the clusters by shifting the centers to a new location
called centroid
9. Centroid is calculated by taking the average position of points
in a cluster
10. Boundaries of the clusters are shifted along with this; Points
are reassigned to other clusters
11. This process of updating clusters is continued till no more
reassigning of points take place

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_data/College_Data",index_col=0)
```

```python
from sklearn.cluster import KMeans
```

```python
df
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop("Private",axis=1))
```

```python
kmeans.cluster_centers_
```

```python
def converter(cluster):
  if cluster == "Yes":
    return 1
  else:
    return 0
```

```python
df["Clusters"]=df["Private"].apply(converter)
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df["Clusters"],kmeans.labels_))
```

```python
print(classification_report(df["Clusters"],kmeans.labels_))
```
