```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
soup
```

```python
names = soup.find_all("div",class_="_4ddWXP")
```

```python
p_names=[]
for i in range(0,len(names)):
  p_names.append(names[i].get_text())
```

```python
p_names
```

```python
rating = soup.find_all("div",class_="_3LWZlK")
```

```python
pr_rating=[]
for i in range(0,len(rating)):
  pr_rating.append(rating[i].get_text())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df["name"] = p_names
```

```python
df["rating"] = pr_rating
```
