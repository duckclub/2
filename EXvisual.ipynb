```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,200)
y = np.arange(201,400)*np.random.randint(100)*x
```

```python
plt.plot(x,y,'r--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("X vs Y")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g+-')
plt.subplot(2,2,2)
plt.plot(x,y,'r+-')
plt.subplot(2,2,3)
plt.plot(x,y,'y+-')
plt.subplot(2,2,4)
plt.plot(x,y,'b+-')
```
